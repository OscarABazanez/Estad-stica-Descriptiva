```
import pandas as pd 
import seaborn as sns

iris = sns.load_dataset('iris')
df = pd.DataFrame(data=iris)
iris.head()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```
# scatterplot 
sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length',hue="species")
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```
# joint plot
markers = {"setosa": "s", "versicolor": "X", "virginia": "X"}

sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length',hue="species", markers=markers)
```

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```
# box plot 
sns.boxplot(data=iris, x = 'species', y = 'sepal_length')
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```
#Barplot
sns.barplot(data=iris, x = 'species', y = 'sepal_length')
```

```
sns.jointplot(data= iris, x= 'sepal_length', y= 'sepal_width' ,hue= 'species' )
```
